================
Lorenz Attractor
================

Аттрактор Лоренца (от англ. to attract — притягивать) ― странный аттрактор, впервые найденный Э. Н. Лоренцем в нелинейной системе обыкновенных дифференциальных уравнений. 
В данной лабораторной работе мы изучаем решение задачи о конвекции в замкнутой петле.


Неизвестные функции x(t), y(t) и z(t) являются гидродинамическими параметрами, характеризующими движение жидкости.
Для простоты, будем считать их средними скоростями движения жидкости в трубке. Здесь x — скорость течения, y — отклонение температуры от средней в точке, отстоящей от нижней точки петли на 90°, z — то же, но в нижней точке. Подведение тепла производится в нижней точке. Параметры b и σ зафиксированны, так как при небольшом изменении этих величин, результат значительно меняется. Таким образом, параметры b и σ в аттракторе Лоренца играют ключевую роль в определении динамики системы и её поведения, влияя на формирование аттракторов и характер динамики системы. 
Еще одно замечательное свойство странных аттракторов, в частности аттрактора Лоренца, – это чувствительность к начальным условиям. Какие бы близкие начальные условия(значения параметров b и σ) мы ни брали, при t→∞ решения будут расходиться, удаляясь друг от друга в фазовом пространстве.

Функция lorenz(xyz, r) принимает в качестве параметров массив координат и значение r. Возвращает массив значений, посчитанный согласно системе уравнений выше.

```python
import matplotlib.pyplot as plt
import numpy as np

s = 10
b = 2.667
dt = 0.01
num_steps = 10000
#arr_r = [0.3, 1.8, 3.7, 10, 16, 24.06, 28, 100]

def lorenz(xyz, r):
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])

```

Функция show_pic(r) принимает параметр r. Создает массив для записи координат, выставляет начальную точку. В цикле высчитывается следующее значение координат x, y, z  в зависимости от посчитанной производной на текущем шаге. В конце происходит отрисовка рисунка в фазовом пространстве.

```python
def show_pic(r):
    xyzs = np.empty((num_steps + 1, 3))  # Need one more for the initial values
    xyzs[0] = (10., 10., 10.)  # Set initial values
    # Step through "time", calculating the partial derivatives at the current point
    # and using them to estimate the next point
    for i in range(num_steps):
        xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i], r) * dt

    # Plot
    ax = plt.figure().add_subplot(projection='3d')

    ax.plot(*xyzs.T, lw=0.5)
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    ax.set_title(f'Lorenz Attractor, r={r}')
    #ax.text(9, -10,0, f'r={r}')
    plt.show()
```

```python
r = 0.3
show_pic(r)
```

```python
r = 1.8
show_pic(r)
```

```python
r = 3.7
show_pic(r)
```

```python
r = 10
show_pic(r)
```

При r = 10 будет происходить ламинарная конвекция. Такому режиму соответствует устойчивый узел решения системы Лоренца.

```python
r = 14
show_pic(r)
```

Если градиент температуры превышает некоторый порог (особо это заметно при переходе от r=14 к r=16), т.е. r увеличивается до сверхбифуркационного значения, то происходит переход от ламинарного движения жидкости к турбулентному. С точки зрения нелинейной динамики это
означает переход от особой точки типа узла, к другому, совершенно необычному режиму, который называется странным аттрактором или аттрактором Лоренца.

```python
r = 16
show_pic(r)
```
